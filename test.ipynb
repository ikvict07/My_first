```python
def recursive_fib(n):
    if n == 1:
        return 1
    elif n == 0:
        return 0
    return recursive_fib(n - 1) + recursive_fib(n - 2)


def recursive_fib_with_cache(n, cache=None):
    if n<2:
        return n
    cache ={}
    if n in cache:
        return cache[n]
    else:
        cache[n] =recursive_fib_with_cache(n-1,cache) + recursive_fib_with_cache(n-2,cache)

    return cache[n]



def fib(n):
    f = [0, 1]
    for i in range(n):
        f[1], f[0] = f[0] + f[1], f[1]
    return f[0]
```
